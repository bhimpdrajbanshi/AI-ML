```python
from tensorflow.keras.preprocessing.text import Tokenizer

# Sample text data
texts = [
    "Nepal is beautiful country.",
    "The dog barks loudly.",
    "Nepali are freindly.",
    "Nepal is beautiful country.",
    "The dog barks loudly.",
    "Nepali are freindly.",
    "Nepal is beautiful country.",
    "The dog barks loudly.",
    "Nepali are freindly."
]

# Create a Tokenizer object
# num_words limits the vocabulary to the top N most frequent words
tokenizer = Tokenizer(num_words=10, oov_token="<unk>")

# Build the vocabulary from the texts
tokenizer.fit_on_texts(texts)

# Convert texts to sequences of integers
sequences = tokenizer.texts_to_sequences(texts)
sequences
```

```python
tokenizer.word_index
```

```python
tokenizer.word_counts
```

```python
tokenizer.document_count
```

```python
from tensorflow import pad
from tensorflow.keras.utils import pad_sequences

sequences = pad_sequences(sequences, padding="pre")
sequences
```
